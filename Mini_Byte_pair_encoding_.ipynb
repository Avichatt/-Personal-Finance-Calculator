```python


# write a def get_status = (words) which taks in the token frequency



# and figure out subtoken cumulative frequency map where each subtoken is comprising of
# 2 continuous letters in the token corpus

words = {
    "l o w": 5,
    "l o w e r": 2,
}

def get_status(words):

    pairs = {}

    for word, freq in words.items():
        symbols = word.split()
        for i in range(len(symbols) - 1):
            pair = (symbols[i], symbols[i+1])
            pairs[pair] = pairs.get(pair, 0) + freq

    return pairs

output = get_status(words)
print(output)

```

```python
#Activity 1: EASY


import re


stopwords = {"the", "is", "of", "and"}

def custom_tokenizer(text):

    text = text.lower()
    text = re.sub("[^a-z0-9]", " ", text)
    tokens = text.split()
    tokens = [t for t in tokens if t not in stopwords]
    return tokens


text = "The price of AI is $1000"
print(custom_tokenizer(text))
```

```python
words = {
    "l o w": 5,
    "l o w e r": 2,
    "n e w e s t": 6
}

def get_status(words):

    pairs = {}

    for word, freq in words.items():
        symbols = word.split()   #Splits the word string into a list of symbols (characters separated by spaces).
        for i in range(len(symbols) - 1):
            pair = (symbols[i], symbols[i+1])   # Creates a tuple representing the adjacent pair.
            pairs[pair] = pairs.get(pair, 0) + freq

    return pairs

output = get_status(words)
print(output)
```

```python
# Activity2: Medium

words = {
    "l o w": 5,
    "l o w e r": 2,
    "n e w e s t": 6
}

def get_stats(vocab):
    pairs = {}
    for word, freq in vocab.items():
        symbols = word.split()
        for i in range(len(symbols) - 1):
            pair = (symbols[i], symbols[i+1])
            pairs[pair] = pairs.get(pair, 0) + freq
    return pairs

def merge_vocab(pair, vocab):
    bigram = " ".join(pair)
    new_vocab = {}
    for word in vocab:
        # replacing the bigram with merged token
        new_word = word.replace(bigram, "".join(pair))
        new_vocab[new_word] = vocab[word]
    return new_vocab


vocab = words.copy()
for i in range(5):
    pairs = get_stats(vocab)
    best_pair = max(pairs, key=pairs.get)   # most frequent pair
    print(f"Iteration {i+1}: best pair = {best_pair}")
    vocab = merge_vocab(best_pair, vocab)
    print("Updated vocab:", vocab)
```
